```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("traffic_accident_data.csv")

# Set seaborn theme
sns.set_theme(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(3, 2, figsize=(15, 14))
fig.suptitle(" Traffic Accident Data Analysis", fontsize=18, weight='bold')

# --- 1. Accidents by Time of Day ---
colors = sns.color_palette("pastel", df['Time_of_Day'].nunique())
df['Time_of_Day'].value_counts().plot(kind='bar', color=colors, ax=axes[0, 0], edgecolor='black')
axes[0, 0].set_title("Accidents by Time of Day")
axes[0, 0].set_xlabel("Time of Day")
axes[0, 0].set_ylabel("Count")

# --- 2. Accidents by Weather Condition ---
colors = sns.color_palette("light:#66c2a5", df['Weather_Condition'].nunique())
df['Weather_Condition'].value_counts().plot(kind='bar', color=colors, ax=axes[0, 1], edgecolor='black')
axes[0, 1].set_title("Accidents by Weather Condition")
axes[0, 1].set_xlabel("Weather")
axes[0, 1].set_ylabel("Count")
axes[0, 1].tick_params(axis='x', rotation=45)

# --- 3. Accidents by Road Condition ---
colors = sns.color_palette("light:#8da0cb", df['Road_Condition'].nunique())
df['Road_Condition'].value_counts().plot(kind='bar', color=colors, ax=axes[1, 0], edgecolor='black')
axes[1, 0].set_title("Accidents by Road Condition")
axes[1, 0].set_xlabel("Road Condition")
axes[1, 0].set_ylabel("Count")

# --- 4. Accident Severity Distribution ---
colors = sns.color_palette("light:#fc8d62", df['Accident_Severity'].nunique())
df['Accident_Severity'].value_counts().sort_index().plot(kind='bar', color=colors, ax=axes[1, 1], edgecolor='black')
axes[1, 1].set_title("Accident Severity (1 = Least, 5 = Most)")
axes[1, 1].set_xlabel("Severity Level")
axes[1, 1].set_ylabel("Count")

# --- 5. Top 10 Locations with Darker Palette ---
top_locations = df['Location'].value_counts().head(10)
dark_colors = sns.color_palette("dark:#4c72b0", len(top_locations))
top_locations.plot(kind='bar', color=dark_colors, ax=axes[2, 0], edgecolor='black')
axes[2, 0].set_title("Top 10 Accident Locations", fontsize=13)
axes[2, 0].set_xlabel("Location")
axes[2, 0].set_ylabel("Count")
axes[2, 0].tick_params(axis='x', rotation=45)

# Hide the unused subplot
axes[2, 1].axis('off')

# Layout adjustment
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
